# 1. Download the data and create a dictionary of results

Since you won't learn about apis, requests, and JSONs until later, here is a function you can use.  It takes a filepath to a JSON and returns a dictionary

```python
def get_dictionary_from_web(filepath):
    
    #download the JSON from 'filepath'
    response = requests.get(filepath)
    
    #transform the JSON into a python dictionary and return the dictionary
    dictionary = json.loads(response.text)
    
    return dictionary
```

Use the link below to return a dictionary of Rick and Morty character information

https://rickandmortyapi.com/api/character 

Examine the keys and values of the dictionary.

```python
characters = get_dictionary_from_web(filepath='https://rickandmortyapi.com/api/character')
characters
```

The API response contains two keys, `'info'` and `'results'`.  `rick_and_morty['info']` is also a dictionary with four keys, `'count'`, `'pages'`, `'next'`, `'prev'`.  Does the list value of `'results':` contain the number of characters described in '`info['count']`?  

If not, check the documentation for the API at https://rickandmortyapi.com/documentation to determine how to get of all the characters.  Use the `get_dictionary_from_web()` function to append all of the characters to a list called  `all_characters`.

Print the length of `all_characters` to make sure it matches the value of `info['count']`

```python
total_pages = 34
all_characters = []
for x in range(total_pages):
    filepath = 'https://rickandmortyapi.com/api/character?page={}'.format(x+1)
    result = get_dictionary_from_web(filepath)
    all_characters.extend(result['results'])
    print(x)
```

# 2. Split the results

Great!  We now have a list of dictionaries with each dictionary containing information about one of the many named characters in Rick and Morty.

Now let's do some do some analysis.  How many characters are alive and how many characters are dead?

Let's split up our characters into 3 dictionaries, `living_characters`, `dead_characters`, `unknown_characters`.  These should be dictionaries of dictionaries.  The keys of the top level dictionary are the characters' names and the values should be the dictionaries of character information for the characters with that name.



```python
# using dictionary comprehension to create 3 dicts based on 'status'
living_characters = {char['name']: char for char in 
                     all_characters if char['status'] == 'Alive'}
dead_characters = {char['name']: char for char in 
                   all_characters if char['status'] == 'Dead'}
unknown_characters = {char['name']: char for char in 
                      all_characters if char['status'] == 'unknown'}
```

```python
# checking dict to make sure all info is included
# note that the 'episode' value has changed since all_characters
living_characters
```

How long are each of each dictionaries of characters?

```python
print(len(living_characters))
print(len(dead_characters))
print(len(unknown_characters))
```
